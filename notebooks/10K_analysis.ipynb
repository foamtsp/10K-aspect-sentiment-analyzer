## 3.1 Data Cleaning

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
```

```python
# Basic info on training set
small_train.info()
```

```python
# Check missing values across all splits
print('=== Missing Values ===')
for name, df in [('Train', small_train), ('Validate', small_validate), ('Test', small_test)]:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if len(missing) == 0:
        print(f'{name}: No missing values')
    else:
        print(f'\n{name}:')
        print(missing)
        print(f'  Missing % of total rows:')
        print((missing / len(df) * 100).round(2))
```

```python
# Check for duplicate rows
for name, df in [('Train', small_train), ('Validate', small_validate), ('Test', small_test)]:
    n_dup = df.duplicated().sum()
    print(f'{name}: {n_dup} duplicate rows ({n_dup/len(df)*100:.2f}%)')
```

```python
# Check for empty / whitespace-only sentences
for name, df in [('Train', small_train), ('Validate', small_validate), ('Test', small_test)]:
    empty = df['sentence'].str.strip().eq('').sum()
    null_sent = df['sentence'].isnull().sum()
    print(f'{name}: {null_sent} null sentences, {empty} empty/whitespace-only sentences')
```

```python
# --- Apply Cleaning ---

def clean_data(df):
    """Apply cleaning steps to a split."""
    df = df.copy()

    # 1. Drop fully duplicated rows
    before = len(df)
    df = df.drop_duplicates()
    print(f'  Dropped {before - len(df)} duplicate rows')

    # 2. Drop rows where sentence is null or empty
    before = len(df)
    df = df.dropna(subset=['sentence'])
    df = df[df['sentence'].str.strip() != '']
    print(f'  Dropped {before - len(df)} null/empty sentence rows')

    # 3. Strip whitespace from sentence
    df['sentence'] = df['sentence'].str.strip()

    # 4. Parse dates
    df['filingDate'] = pd.to_datetime(df['filingDate'], errors='coerce')
    df['reportDate'] = pd.to_datetime(df['reportDate'], errors='coerce')

    # 5. Normalize label columns
    for col in ['label_1d', 'label_5d', 'label_30d']:
        df[col] = df[col].str.strip().str.lower()

    df = df.reset_index(drop=True)
    return df

print('Cleaning Train:')
small_train = clean_data(small_train)
print('Cleaning Validate:')
small_validate = clean_data(small_validate)
print('Cleaning Test:')
small_test = clean_data(small_test)

print(f'\nAfter cleaning — Train: {small_train.shape}, Validate: {small_validate.shape}, Test: {small_test.shape}')
```

## 3.4 Initial Modeling (Baseline)

Simple **TF-IDF + Logistic Regression** and **Engineered Features + Random Forest** baselines to check if text carries predictive signal for `label_30d`.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score

# Prepare data
X_train_text = small_train['sentence']
y_train = (small_train['label_30d'] == 'positive').astype(int)

X_val_text = small_validate['sentence'].fillna('')
y_val = (small_validate['label_30d'].str.strip().str.lower() == 'positive').astype(int)

X_test_text = small_test['sentence'].fillna('')
y_test = (small_test['label_30d'].str.strip().str.lower() == 'positive').astype(int)

print(f'Train: {len(X_train_text)}, Validate: {len(X_val_text)}, Test: {len(X_test_text)}')
print(f'Train label balance: {y_train.mean()*100:.1f}% positive')
```

```python
# TF-IDF + Logistic Regression
tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1, 2),
                         min_df=3, max_df=0.95, sublinear_tf=True)

X_train_tfidf = tfidf.fit_transform(X_train_text)
X_val_tfidf = tfidf.transform(X_val_text)
X_test_tfidf = tfidf.transform(X_test_text)

print(f'TF-IDF matrix shape: {X_train_tfidf.shape}')

lr = LogisticRegression(max_iter=1000, C=1.0, random_state=42, n_jobs=-1)
lr.fit(X_train_tfidf, y_train)

print('\n=== Validation Set ===')
y_val_pred = lr.predict(X_val_tfidf)
y_val_proba = lr.predict_proba(X_val_tfidf)[:, 1]
print(classification_report(y_val, y_val_pred, target_names=['negative', 'positive']))
print(f'ROC AUC: {roc_auc_score(y_val, y_val_proba):.4f}')

print('\n=== Test Set ===')
y_test_pred = lr.predict(X_test_tfidf)
y_test_proba = lr.predict_proba(X_test_tfidf)[:, 1]
print(classification_report(y_test, y_test_pred, target_names=['negative', 'positive']))
print(f'ROC AUC: {roc_auc_score(y_test, y_test_proba):.4f}')
```

```python
# Confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for ax, y_true, y_pred, title in zip(axes,
    [y_val, y_test], [y_val_pred, y_test_pred],
    ['Validation', 'Test']):
    cm = confusion_matrix(y_true, y_pred)
    ConfusionMatrixDisplay(cm, display_labels=['negative', 'positive']).plot(ax=ax, cmap='Blues')
    ax.set_title(f'{title} Set Confusion Matrix', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Top predictive words
feature_names = tfidf.get_feature_names_out()
coefs = lr.coef_[0]

top_positive_idx = np.argsort(coefs)[-20:]
top_negative_idx = np.argsort(coefs)[:20]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].barh(range(20), coefs[top_positive_idx], color='#2ecc71', edgecolor='black', linewidth=0.3)
axes[0].set_yticks(range(20))
axes[0].set_yticklabels(feature_names[top_positive_idx])
axes[0].set_title('Top 20 Words Predicting POSITIVE', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Coefficient')

axes[1].barh(range(20), coefs[top_negative_idx], color='#e74c3c', edgecolor='black', linewidth=0.3)
axes[1].set_yticks(range(20))
axes[1].set_yticklabels(feature_names[top_negative_idx])
axes[1].set_title('Top 20 Words Predicting NEGATIVE', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Coefficient')

plt.suptitle('Most Predictive TF-IDF Features (Logistic Regression)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### Baseline 2: Random Forest on Engineered Features

```python
from sklearn.ensemble import RandomForestClassifier

def engineer_features(df):
    features = pd.DataFrame()
    features['sentence_len'] = df['sentence'].str.len()
    features['word_count'] = df['sentence'].str.split().str.len()
    features['avg_word_len'] = features['sentence_len'] / (features['word_count'] + 1)
    features['num_commas'] = df['sentence'].str.count(',')
    features['num_periods'] = df['sentence'].str.count(r'\.')
    features['num_digits'] = df['sentence'].str.count(r'\d')
    features['pct_digits'] = features['num_digits'] / (features['sentence_len'] + 1)
    features['num_dollar'] = df['sentence'].str.count(r'\$')
    features['num_percent'] = df['sentence'].str.count(r'%')
    features['has_numbers'] = (features['num_digits'] > 0).astype(int)
    section_map = {'section_1': 0, 'section_1A': 1, 'section_7': 2, 'section_7A': 3}
    features['section_code'] = df['section'].map(section_map).fillna(-1).astype(int)
    features['sic'] = pd.to_numeric(df['sic'], errors='coerce').fillna(0).astype(int)
    return features

X_train_feat = engineer_features(small_train)
X_val_feat = engineer_features(small_validate)
X_test_feat = engineer_features(small_test)

rf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1)
rf.fit(X_train_feat, y_train)

print('=== Random Forest — Validation Set ===')
y_val_pred_rf = rf.predict(X_val_feat)
y_val_proba_rf = rf.predict_proba(X_val_feat)[:, 1]
print(classification_report(y_val, y_val_pred_rf, target_names=['negative', 'positive']))
print(f'ROC AUC: {roc_auc_score(y_val, y_val_proba_rf):.4f}')

print('\n=== Random Forest — Test Set ===')
y_test_pred_rf = rf.predict(X_test_feat)
y_test_proba_rf = rf.predict_proba(X_test_feat)[:, 1]
print(classification_report(y_test, y_test_pred_rf, target_names=['negative', 'positive']))
print(f'ROC AUC: {roc_auc_score(y_test, y_test_proba_rf):.4f}')
```

```python
# Feature importance
importances = rf.feature_importances_
feat_names = X_train_feat.columns
sorted_idx = np.argsort(importances)

fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(feat_names[sorted_idx], importances[sorted_idx], color='#3498db', edgecolor='black', linewidth=0.3)
ax.set_title('Random Forest Feature Importances', fontsize=14, fontweight='bold')
ax.set_xlabel('Importance')
plt.tight_layout()
plt.show()
```

### Model Comparison Summary

2. **Class Balance**: Check the label distribution above — if imbalanced, we may need stratified sampling or class weights in later models.
3. **Sections**: MD&A (Section 7) and Risk Factors (Section 1A) typically contain the most informative text for predicting price direction.
4. **Baseline Performance**: The TF-IDF + Logistic Regression baseline gives us a floor to beat. Any NLP model (BERT, FinBERT) should improve on this.
5. **Next Steps**: Aggregate sentences to filing level, apply sentiment models (FinBERT), and train on the large dataset.
